```python
import chardet
import pandas as pd


# 파일의 인코딩 추정
with open('/content/jobkorea_cv_2_finetuning.csv', 'rb') as f:
    result = chardet.detect(f.read())

# 추정된 인코딩 출력
print(result)

# 추정된 인코딩으로 파일 읽기
df = pd.read_csv('/content/jobkorea_cv_2_finetuning.csv', encoding=result['encoding'])
df.head()
```

```python
df = df.dropna(subset=['answer1'])
len(df) #answer1 이 결측값인 행 제거
```

```python
# 데이터 전처리 함수
'''def preprocess_function(examples):
    inputs = examples['answer1']
    targets = examples['finetuning']
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=512, truncation=True, padding="max_length")
    model_inputs['labels'] = labels['input_ids']
    return model_inputs'''

def preprocess_function(examples):
    inputs = examples['answer1']
    targets = examples['finetuning']
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=512, truncation=True, padding="max_length")
    model_inputs['labels'] = labels['input_ids']
    return model_inputs
```

```python
from sklearn.model_selection import train_test_split
from datasets import Dataset
from transformers import Trainer, TrainingArguments, AutoTokenizer, AutoModelForCausalLM

train_df, eval_df = train_test_split(df, test_size=0.3)
train_dataset = Dataset.from_pandas(train_df)
eval_dataset = Dataset.from_pandas(eval_df)
```

```python
#tokenizer.add_special_tokens({'pad_token': '[PAD]'})
```

```python
tokenized_train_dataset = train_dataset.map(preprocess_function, batched=True)
tokenized_eval_dataset = eval_dataset.map(preprocess_function, batched=True)
```

```python
training_args = TrainingArguments(
    output_dir="/content",
    evaluation_strategy="epoch",
    save_strategy="epoch",  # save_strategy를 eval_strategy와 동일하게 설정
    learning_rate=3e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=10,
    weight_decay=0.01,
    fp16=True,  # 혼합 정밀도 학습 활성화
    logging_steps=50,
    save_total_limit=3,  # 체크포인트 제한
    load_best_model_at_end=True,  # 최적 모델 로드
    metric_for_best_model="eval_loss",  # 최적 모델 기준
    greater_is_better=False  # 손실 값은 낮을수록 좋기 때문에 False로 설정
)
```

```python
from transformers import Trainer, TrainingArguments, EarlyStoppingCallback

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_eval_dataset,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
)
```

```python
trainer.train()
```
