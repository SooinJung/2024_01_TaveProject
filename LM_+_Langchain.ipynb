## 허깅페이스 파이프라인으로 랭체인에 올리기
이렇게 해야 된다. 왜 위의 방법으로 하면 안되는지 잘 모르겠다...

```python
from langchain_community.llms.huggingface_pipeline import HuggingFacePipeline

# HuggingFace 모델을 다운로드 받습니다.
hf = HuggingFacePipeline.from_model_id(
    model_id="beomi/llama-2-ko-7b",  # 사용할 모델의 ID를 지정합니다.
    task="text-generation",  # 수행할 작업을 지정합니다. 여기서는 텍스트 생성입니다.
    # 파이프라인에 전달할 추가 인자를 설정합니다. 여기서는 생성할 최대 토큰 수를 10으로 제한합니다.
    pipeline_kwargs={"max_new_tokens": 512},
    device = 0, # 첫번째 GPU 사용하게 설정
)
```

```python
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_huggingface import HuggingFacePipeline
from transformers import pipeline

# 프롬프트 템플릿 정의
template = """You are an expert AI interviewer.
Make a question in korean related to the user's resume.

#resume
{resume}

#Question: """
prompt = PromptTemplate(template=template, input_variables=["resume"])

# 체인 생성
chain = prompt | hf | StrOutputParser()

# 이력서 텍스트 정의
resume = '''보건관리학과와 정보통계학을 복수 전공하면서, 학부시절부터 이를 융합한 직무에 관심이 많았습니다. 졸업 이후에는 연세의료원 산학협력단 소속 연구원으로 재직하며 디지털헬스, 의료빅데이터에 관한 업무들을 하였습니다. 의료 데이터를 전처리하여 이를 waterfall plot, spider plot, survival plot 등 R 프로그램으로 구현이 용이한 것들을 파이썬으로 구현하였고, 개발자와 협엽하여 이를 django로 연결시킬 수 있도록 하였습니다. 또한 세브란스 병원의 데이터 플랫폼인 Yonsei Cancer Data Library 구축 프로젝트의 일환으로 QC 파트에 참여하여 MS-SQL을 이용해 임상지식에 기반하여 틀린 로직들을 수정하는 업무를 하였습니다.  이러한 저의 업무 경험과 더불어 보건학적 지식과 빅데이터 지식을 활용하여 더존비즈온의 의료빅데이터 사업기획 및 관리 업무에 기여하고 싶어 지원하게 되었습니다. 입사하게 된다면, 새로 접하는 기술이나 지식이 있을 때 주체적인 태도를 가지고 공부하여 좋은 방안을 제시할 수 있도록 할 것 입니다. 끊임없이 변화하는 기술 트렌드를 따라갈 수 있도록 Lancet Digital Health와 같은 저널들을 자주 접하도록 하겠습니다. 또한 학회도 참여하여 국내 최신 트렌드와 기술을 공부하여 더존비즈온의 기술에 적용할 수 있는 아이디어를 제시할 수 있도록 할 것입니다.'''

# 체인을 호출하여 질문에 대한 답변 생성 및 출력
result = chain.invoke({"resume": resume})
print(result)
```
