```python
import os
#import torch
#import numpy as np
#from transformers import AutoTokenizer, AutoModelForCausalLM
#from langchain.embeddings import OpenAIEmbeddings  # OpenAI 임베딩 사용
from langchain.llms import OpenAI
from langchain import PromptTemplate
from langchain.chains import LLMChain

os.environ["OPENAI_API_KEY"] = "" # 직접 입력

# 질문 템플릿 형식 정의
template = '''
You are an expert AI interviewer.
Use the following resume to make a question in korean: {resume}
Please give only one question and don't add any explanations.
And please give a specific question. And the question should be related to the user's resume.
And please do not ask the same question again before that you made to user.
Focus on generating a variety of question. And please make a detail question which is based on the skills which user mentioned.
'''

# 템플릿 완성
prompt = PromptTemplate(template=template, input_variables=['resume'])

# 연결된 체인(Chain)객체 생성
llm = OpenAI()
llm_chain = LLMChain(prompt=prompt, llm=llm)

resume = input("사용자의 자소서를 입력해주세요: ")

for i in range(5):
    # 사용자 답변 받기
    question = llm_chain.run(resume=resume)
    print(f"질문 {i+1}: {question}")
    answer = input(f"답변 {i+1}: \n")

    # 답변 내용을 다음 질문을 위한 resume에 추가 (for 꼬리 질문)
    resume += f" {answer}"
```
