```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "klue/roberta-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, is_decoder=True)
```

```python
import chardet
import pandas as pd

# 파일의 인코딩 감지
with open('/content/jobkorea_cv_1.csv', 'rb') as f:
    result = chardet.detect(f.read())


# 감지된 인코딩 사용하여 CSV 파일 읽기
encoding = result['encoding']
df = pd.read_csv('/content/jobkorea_cv_1.csv', encoding=encoding)

# 데이터 프레임 출력
df
```

NAN 값이 좀 들어가있는 행 삭제 (전처리)

```python
df.drop(index=3, inplace=True)
df
```

```python
# input 컬럼과 target 컬럼을 병합
inputs = df[['answer1', 'answer2', 'answer3']].fillna('')
inputs = inputs.apply(lambda x: ' '.join(x), axis=1)

targets = df[['question1', 'question2', 'question3']].fillna('')
targets = targets.apply(lambda x: ' '.join(x), axis=1)

# 새로운 데이터 프레임 생성
processed_df = pd.DataFrame({'input': inputs, 'target': targets})

# 데이터 프레임 출력
processed_df
```

```python
# drop=True는 기존 index를 drop 한다는 것
processed_df.reset_index(drop=True, inplace=True)
processed_df
```

```python
from sklearn.model_selection import train_test_split
from datasets import Dataset
train_df, eval_df = train_test_split(processed_df, test_size=0.3)
train_dataset = Dataset.from_pandas(train_df)
eval_dataset = Dataset.from_pandas(eval_df)
from transformers import Trainer, TrainingArguments, AutoTokenizer, AutoModelForCausalLM
```

```python
# 데이터 전처리 함수
def preprocess_function(examples):
    inputs = examples['input']
    targets = examples['target']
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=512, truncation=True, padding="max_length")
    model_inputs['labels'] = labels['input_ids']
    return model_inputs
```

```python
tokenized_train_dataset = train_dataset.map(preprocess_function, batched=True)
tokenized_eval_dataset = eval_dataset.map(preprocess_function, batched=True)
```

```python
training_args = TrainingArguments(
    output_dir="/content",
    evaluation_strategy="epoch",
    save_strategy="epoch",  # save_strategy를 eval_strategy와 동일하게 설정
    learning_rate=3e-5,
    per_device_train_batch_size=16, # 메모리가 허용하는 한 최대 크기
    per_device_eval_batch_size=16,
    num_train_epochs=30,
    weight_decay=0.01,
    fp16=True,  # 혼합 정밀도 학습 활성화
    logging_steps=50, # 로그를 자주 확인하여 학습 과정 모니터링
    save_total_limit=3,  # 체크포인트 너무 자주 저장하지 않도록
    load_best_model_at_end=True,  # 최적 모델 로드
    metric_for_best_model="eval_loss",  # 최적 모델 기준
    greater_is_better=False  # 손실 값은 낮을수록 좋기 때문에 False로 설정
)
```

```python
from transformers import Trainer, TrainingArguments, EarlyStoppingCallback

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,
    eval_dataset=tokenized_eval_dataset,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)] # Early Stopping
)
```

```python
trainer.train()
```
